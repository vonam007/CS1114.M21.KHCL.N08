# Thng k train

```python
import glob, os
os.chdir('./data_final/')
txt = []
c_0 = 0
c_1 = 0
c_2 = 0
c_3 = 0
c_4 = 0
c_5 = 0
c_6 = 0
c_7 = 0
c_8 = 0
total_img = 0
for file in glob.glob("*.txt"):
  txt.append(file)
total_img = len(txt)
for i in range(len(txt)):
    fl = open(txt[i], 'r')
    data = fl.readlines()
    fl.close()

    for dt in data:
        c, x_min, y_min, x_max, y_max = map(str, dt.split(' '))
        c = int(c)
        if c == 0:
          c_0+=1
        elif c == 1:
          c_1+=1
        elif c == 2:
          c_2+=1
        elif c==3:
          c_3+=1
        elif c==4:
          c_4 += 1
        elif c==5:
          c_5+=1
        elif c == 6:
          c_6 +=1
        elif c == 7:
          c_7 += 1   
        elif c == 8:
          c_8 += 1     

print('Total_img: ',total_img)
print('No entry: ',c_0)
print('No parking / waiting: ',c_1)
print('No turning: ',c_2)
print('Max speed: ',c_3)
print('Other prohibition signs: ',c_4)
print('Warning: ',c_5)
print('Mandatory: ',c_6)
print('Indication: ',c_7)
print('Other sign: ',c_8)
```
