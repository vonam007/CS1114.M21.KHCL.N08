```python
# Seprate data train - test
import os
import glob2
import numpy as np

all_files = []
for ext in ["*.png", "*.jpeg", "*.jpg"]:
  images = glob2.glob(os.path.join("data/data_final/", ext))
  all_files += images

# rand_idx = np.random.randint(0, len(all_files), 200)

# Create train.txt
with open("data/train.txt", "w") as f:
  for idx in np.arange(len(all_files)):
    f.write(all_files[idx]+'\n')
```

```python
# Seprate data train - test
import os
import glob2
import numpy as np

all_files = []
for ext in ["*.png", "*.jpeg", "*.jpg"]:
  images = glob2.glob(os.path.join("data/data_test/", ext))
  all_files += images


# Create valid.txt
with open("data/valid.txt", "w") as f:
  for idx in np.arange(len(all_files)):
    f.write(all_files[idx]+'\n')

```
