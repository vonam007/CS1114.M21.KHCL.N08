```python
net = cv2.dnn.readNet("/content/drive/MyDrive/CS114/yolov4/darknet/backup/yolov4-custom_best.weights", "/content/drive/MyDrive/CS114/yolov4/darknet/cfg/yolov4-custom.cfg")
# output_path = os.path.join("/content/drive/MyDrive/CS114/data_update_1/data_test", "*.jpg")
# Name custom object
classesFile = "/content/drive/MyDrive/CS114/yolov4/darknet/data/obj.names"
classes = None
```

```python
with open(classesFile, 'rt') as f:
    classes = f.read().rstrip('\n').split('\n')
layer_names = net.getLayerNames()
output_layers = [layer_names[i[0] - 1] for i in net.getUnconnectedOutLayers()]
colors = np.random.uniform(0, 255, size=(len(classes), 3))
```

```python
def convert(size, box):
    dw = 1./(size[0])
    dh = 1./(size[1])
    x = (box[0] + box[1])/2.0 - 1
    y = (box[2] + box[3])/2.0 - 1
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return (x,y,w,h)
```

```python
def detect_image(img):
    images=img.split('/')[-1]
    class_name = str()
    #print(images)
    output_path = os.path.join("results", images)
    img=cv2.imread(img)
    height, width, channels = img.shape

    blob = cv2.dnn.blobFromImage(img, 0.00392, (416, 416), (0, 0, 0), True, crop=False)

    net.setInput(blob)
    outs = net.forward(output_layers)

    class_ids = []
    confidences = []
    boxes = []
    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5:
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width)
                h = int(detection[3] * height)

                x = int(center_x - w / 2)
                y = int(center_y - h / 2)

                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)
                
    indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)
    font = cv2.FONT_HERSHEY_PLAIN
    for i in range(len(boxes)):
        if i in indexes:
            x, y, w, h = boxes[i]
            # box_new=[x,x+w,y,y+h]#bbox on pascalVOC
            # new_bbox=convert((width,height),box_new)#bbox on yolo
            new_bbox = [x,y,x+w,y+h]
            print(class_ids[i],confidences[i],new_bbox[0],new_bbox[1],new_bbox[2],new_bbox[3])
            if class_ids[i] == 0:
              class_name = "No_entry"
            elif class_ids[i] == 1:
              class_name = "No_parking_waiting"
            elif class_ids[i] == 2:
              class_name = "No_turning"
            elif class_ids[i] == 3:
              class_name = "Max_speed"
            elif class_ids[i] == 4:
              class_name = "Other_prohibition_signs"
            elif class_ids[i] == 5:
              class_name = "Warning"
            elif class_ids[i] == 6:
              class_name = "Mandatory"
            elif class_ids[i] == 7:
              class_name = "Indication"
            elif class_ids[i] == 8:
              class_name = "Other_sign"

            name=images.split(".")[0]
            name="detection-results/"+name+".txt"
            with open(name,'a',encoding = 'utf-8') as f:
                f.write(str(class_name) + " ")
                f.write(str(confidences[i]) + " ")   
                f.write(str(new_bbox[0]) + " ")
                f.write(str(new_bbox[1]) + " ")
                f.write(str(new_bbox[2]) + " ")
                f.write(str(new_bbox[3]))
                f.write("\n") 

    return img, output_path
```

```python
def create_predictor(model_config, model_weights, threshold, num_of_class):
    cfg = get_cfg()
    cfg.merge_from_file(model_config)
    cfg.MODEL.WEIGHTS = model_weights
    cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = threshold
    cfg.MODEL.ROI_HEADS.NUM_CLASSES = num_of_class
    predictor = DefaultPredictor(cfg)
    return predictor
```

```python
def create_submission(folder_dir, predictor):
    # file_names = os.listdir(folder_dir)
    file_names = [f for f in os.listdir(folder_dir) if f.endswith('.jpg')]
    file_names.sort()
    for file_name in file_names:
        # print(file_name)
        image_path = os.path.join(folder_dir, file_name)
        image = cv2.imread(image_path)
        output = predictor(image)
        boxes = output['instances'].pred_boxes
        scores = output['instances'].scores
        classes = output['instances'].pred_classes
        for i in range(len(classes)):
            class_id = int(classes[i])
            # No entry
            # No parking / waiting
            # No turning
            # Max speed
            # Other prohibition signs
            # Warning
            # Mandatory
            # Indication
            # Other sign
            if class_id == 0:
              class_name = "No_entry"
            elif class_id == 1:
              class_name = "No_parking_waiting"
            elif class_id == 2:
              class_name = "No_turning"
            elif class_id == 3:
              class_name = "Max_speed"
            elif class_id == 4:
              class_name = "Other_prohibition_signs"
            elif class_id == 5:
              class_name = "Warning"
            elif class_id == 6:
              class_name = "Mandatory"
            elif class_id == 7:
              class_name = "Indication"
            elif class_id == 8:
              class_name = "Other_sign"
            if (scores[i] > 0.5):
                for j in boxes[i]:
                    confident = np.array(scores[i].cpu())
                    x_min = int(j[0])
                    y_min = int(j[1])
                    x_max = int(j[2])
                    y_max = int(j[3])
                    # content = str(file_name) + ' ' + str(class_id) + ' ' + str(x_min) + ' ' + str(y_min) + ' ' + str(x_max) + ' ' + str(y_max) + '\n'
                    content = str(class_name) + ' '+str(confident)+' '+ str(x_min) + ' ' + str(y_min) + ' ' + str(x_max) + ' ' + str(y_max) + '\n'
                    txt_name = file_name
                    txt_name = txt_name.replace('jpg','txt')
                    f = open(txt_name, "a")
                    f.write(content)
                    f.close()
```
